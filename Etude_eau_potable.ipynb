## MortalityRateAttributedToWater

```python
MortalityRateAttributedToWater.pivot(index = ['Country','Year'], columns='Granularity', values=['Mortality rate attributed to exposure to unsafe WASH services', 'WASH deaths'])
```

### Wash_deaths

```python
Wash_deaths = MortalityRateAttributedToWater.drop('Mortality rate attributed to exposure to unsafe WASH services', axis=1)
```
